 ARRAY CREATION METHODS

```python
# ==============================
# NUMPY TOPIC 4 : ARRAY CREATION METHODS
# ==============================

import numpy as np

print("NumPy Array Creation Methods\n")

# ==============================
# 1. zeros()
# ==============================

zeros_array = np.zeros((2,3)) # np.zeros() creates an array filled with zeros. The argument is a tuple that specifies the shape of the array (rows, columns)
print("1. Zeros Array (2x3)")
print(zeros_array)
print()

# ==============================
# 2. ones()
# ==============================

ones_array = np.ones((3,3)) # np.ones() creates an array filled with ones. The argument is a tuple that specifies the shape of the array (rows, columns)
print("2. Ones Array (3x3)")
print(ones_array)
print()

# ==============================
# 3. empty()
# ==============================

empty_array = np.empty((2,2)) # np.empty() creates an array with uninitialized values. The argument is a tuple that specifies the shape of the array (rows, columns)
print("3. Empty Array (2x2)")
print(empty_array)
print()

# ==============================
# 4. full()
# ==============================

full_array = np.full((2,3), 7) # np.full() creates an array filled with a specified value. The first argument is a tuple that specifies the shape of the array (rows, columns), and the second argument is the value to fill the array with.
print("4. Full Array with value 7")
print(full_array)
print()

# ==============================
# 5. arange()
# ==============================

arange_array = np.arange(0,10,2) # np.arange() creates an array with a range of values. The first argument is the start value, the second argument is the end value (exclusive), and the third argument is the step size.
print("5. Arange Array (0 to 10 step 2)")
print(arange_array)
print()

# ==============================
# 6. linespace()
# ==============================

linspace_array = np.linspace(0,1,5) # np.linspace() creates an array of evenly spaced values between a specified start and end. The first argument is the start value, the second argument is the end value, and the third argument is the number of values to generate.
print("6. Linspace Array (5 values between 0 and 1)")
print(linspace_array)
print()

# ==============================
# 7. eye()
# ==============================

eye_array = np.eye(3)
print("7. Identity Matrix using eye()")
print(eye_array)
print()

# ==============================
# 8. identity()
# ==============================

identity_array = np.identity(4)
print("8. Identity Matrix using identity()")
print(identity_array)
print()

print("NumPy Array Creation Methods Completed!")
```

RESHAPING ARRAY

```python
# ==============================
# NUMPY TOPIC 8 : RESHAPING ARRAYS
# ==============================

import numpy as np

print("NumPy Reshaping\n")

# ==============================
# Create original array
# ==============================

arr = np.arange(1,13)

print("Original Array:")
print(arr)

# ==============================
# 1️ reshape()
# ==============================

reshaped = arr.reshape(3,4)

print("\n1. Reshape to 3x4")
print(reshaped)

# ==============================
# 2️ reshape to 2x6
# ==============================

reshaped2 = arr.reshape(2,6)

print("\n2. Reshape to 2x6")
print(reshaped2)

# ==============================
# 3️ flatten()
# ==============================

flat = reshaped.flatten()

print("\n3. Flatten Array (convert to 1D)")
print(flat)

# ==============================
# 4️ ravel()
# ==============================

ravel_arr = reshaped.ravel()

print("\n4. Ravel Array (also converts to 1D)")
print(ravel_arr)

# ==============================
# 5️ resize()
# ==============================

arr_resize = np.array([1,2,3,4])
arr_resize.resize(2,2)

print("\n5. Resize Array to 2x2")
print(arr_resize)

print("\nNumPy Reshaping Topic Completed!")
```

ARRAY ITERATION


```python
# ==============================
# NUMPY TOPIC 9 : ARRAY ITERATION
# ==============================

import numpy as np

print("NumPy Array Iteration\n")

# ==============================
# Create arrays
# ==============================

arr1 = np.array([10,20,30,40])
arr2 = np.array([[1,2,3],
                 [4,5,6]])

# ==============================
# 1️ Iterate 1D Array
# ==============================

print("1. Iterating 1D Array")

for x in arr1:
    print(x)

# ==============================
# 2️ Iterate 2D Array
# ==============================

print("\n2. Iterating 2D Array")

for row in arr2:
    print(row)

# ==============================
# 3️ Iterate each element in 2D
# ==============================

print("\n3. Iterating each element")

for row in arr2:
    for element in row:
        print(element)

# ==============================
# 4️ Using nditer()
# ==============================

print("\n4. Using nditer()")

for x in np.nditer(arr2):
    print(x)

# ==============================
# 5️ Using ndenumerate()
# ==============================

print("\n5. Using ndenumerate()")

for index, value in np.ndenumerate(arr2):
    print(index, value)

print("\nNumPy Iteration Topic Completed!")
```

ARRAY JOINING

```python
# ==============================
# NUMPY TOPIC 10 : ARRAY JOINING
# ==============================

import numpy as np

print("NumPy Array Joining\n")

# Create arrays
arr1 = np.array([1,2,3])
arr2 = np.array([4,5,6])

print("Array 1:", arr1)
print("Array 2:", arr2)

# ==============================
# 1️ concatenate()
# ==============================

concat = np.concatenate((arr1, arr2))

print("\n1. concatenate()")
print(concat)

# ==============================
# 2️ stack()
# ==============================

stack_arr = np.stack((arr1, arr2))

print("\n2. stack()")
print(stack_arr)

# ==============================
# 3️ hstack()
# ==============================

hstack_arr = np.hstack((arr1, arr2))

print("\n3. hstack()")
print(hstack_arr)

# ==============================
# 4️ vstack()
# ==============================

vstack_arr = np.vstack((arr1, arr2))

print("\n4. vstack()")
print(vstack_arr)

# ==============================
# 5️ dstack()
# ==============================

dstack_arr = np.dstack((arr1, arr2))

print("\n5. dstack()")
print(dstack_arr)

print("\nNumPy Array Joining Topic Completed!")
```

ARRAY SPLITING

COPY VS VIEW

```python
# ==============================
# NUMPY TOPIC 12 : COPY vs VIEW
# ==============================

import numpy as np

print("NumPy Copy vs View\n")

# Create original array
arr = np.array([10,20,30,40])

print("Original Array:", arr)

# ==============================
# 1️ Copy
# ==============================

copy_arr = arr.copy()

print("\n1. Copy Array")
print("Copy:", copy_arr)

# Modify original
arr[0] = 100

print("\nAfter modifying original array")
print("Original:", arr)
print("Copy:", copy_arr)

# ==============================
# 2️ View
# ==============================

arr2 = np.array([1,2,3,4])

view_arr = arr2.view()

print("\n2. View Array")
print("Original:", arr2)
print("View:", view_arr)

# Modify original
arr2[0] = 99

print("\nAfter modifying original array")
print("Original:", arr2)
print("View:", view_arr)

# ==============================
# 3️ Check base
# ==============================

print("\n3. Checking base property")

print("Copy base:", copy_arr.base)
print("View base:", view_arr.base)

print("\nNumPy Copy vs View Topic Completed!")
```

SORTING & SEARCHING

```python
# ==============================
# NUMPY TOPIC 18 : SORTING & SEARCHING
# ==============================

import numpy as np

print("NumPy Sorting and Searching\n")

# Create array
arr = np.array([50,10,40,20,30])

print("Original Array:", arr)

# ==============================
# 1️ Sorting
# ==============================

sorted_arr = np.sort(arr)

print("\n1. Sorted Array using np.sort()")
print(sorted_arr)

# ==============================
# 2️ argsort()
# ==============================

index_sorted = np.argsort(arr)

print("\n2. argsort() (indices of sorted array)")
print(index_sorted)

# ==============================
# 3️ where()
# ==============================

where_result = np.where(arr == 20)

print("\n3. where() (find index of value 20)")
print(where_result)

# ==============================
# 4️ searchsorted()
# ==============================

sorted_array = np.array([10,20,30,40,50])

position = np.searchsorted(sorted_array, 35)

print("\n4. searchsorted() (position to insert 35)")
print(position)

print("\nNumPy Sorting & Searching Completed!")
```
