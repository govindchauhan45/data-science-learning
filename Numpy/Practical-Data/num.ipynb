Student Marks Data Analysis System

```python
# ============================================
# ADVANCED NUMPY PROJECT
# Student Marks Data Analysis System
# Covers Most NumPy Topics
# ============================================

import numpy as np

print("===== STUDENT DATA ANALYSIS PROJECT =====")

# --------------------------------------------
# 1. Create Student Data (Random)
# --------------------------------------------

np.random.seed(0)

students = 10
subjects = 5

marks = np.random.randint(40,100,(students,subjects))

print("\nStudent Marks Matrix")
print(marks)

# --------------------------------------------
# 2. Array Shape & Size
# --------------------------------------------

print("\nShape:",marks.shape)
print("Total Elements:",marks.size)
print("Dimensions:",marks.ndim)

# --------------------------------------------
# 3. Indexing & Slicing
# --------------------------------------------

print("\nMarks of First Student")
print(marks[0])

print("\nMarks of First Subject for All Students")
print(marks[:,0])

print("\nFirst 3 Students Data")
print(marks[:3])

# --------------------------------------------
# 4. Mathematical Operations
# --------------------------------------------

print("\nAdd 5 Grace Marks")
grace_marks = marks + 5
print(grace_marks)

# --------------------------------------------
# 5. Broadcasting Example
# --------------------------------------------

bonus = np.array([2,3,4,5,6])

print("\nBroadcasting Bonus Marks")
print(marks + bonus)

# --------------------------------------------
# 6. Statistical Analysis
# --------------------------------------------

print("\nAverage Marks Per Student")
print(np.mean(marks,axis=1))

print("\nHighest Marks Per Subject")
print(np.max(marks,axis=0))

print("\nLowest Marks Per Subject")
print(np.min(marks,axis=0))

print("\nStandard Deviation")
print(np.std(marks))

# --------------------------------------------
# 7. Reshape Example
# --------------------------------------------

flat = marks.reshape(50)

print("\nReshaped to 1D Array")
print(flat)

# --------------------------------------------
# 8. Filtering (Students with marks > 90)
# --------------------------------------------

print("\nMarks Greater Than 90")
print(marks[marks>90])

# --------------------------------------------
# 9. Sorting
# --------------------------------------------

print("\nSorted Marks of First Student")
print(np.sort(marks[0]))

# --------------------------------------------
# 10. Argsort (Ranking Students by Avg)
# --------------------------------------------

avg = np.mean(marks,axis=1)

rank = np.argsort(avg)

print("\nStudent Ranking Index")
print(rank)

# --------------------------------------------
# 11. Searching
# --------------------------------------------

print("\nIndex Where Marks > 95")
print(np.where(marks>95))

# --------------------------------------------
# 12. Stacking Arrays
# --------------------------------------------

extra_data = np.random.randint(50,90,(10,5))

combined = np.vstack((marks,extra_data))

print("\nCombined Dataset Shape")
print(combined.shape)

# --------------------------------------------
# 13. Matrix Operations
# --------------------------------------------

A = np.array([[1,2],[3,4]])
B = np.array([[5,6],[7,8]])

print("\nMatrix Multiplication")
print(np.dot(A,B))

# --------------------------------------------
# 14. Unique Values
# --------------------------------------------

print("\nUnique Marks")
print(np.unique(marks))

# --------------------------------------------
# 15. Percentile Analysis
# --------------------------------------------

print("\n90th Percentile Marks")
print(np.percentile(marks,90))

print("\n===== ANALYSIS COMPLETE =====")
```
